```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ppo_log_txt = open('/workspace/V2/output/ppo/ppo_log.txt', 'r')
ddqn_log_txt = open('/workspace/V2/output/ddqn/ddqn_log.txt', 'r')
```

```python
def make_reward_list(file_name):
    reward_list = []
    lines = file_name.readlines()
    for i in range(len(lines)):
        reward_list.append(float(lines[i].split(' ')[-1][:-2]))
    
    return reward_list
```

```python
ppo_reward_list = make_reward_list(ppo_log_txt)
ddqn_reward_list = make_reward_list(ddqn_log_txt)
```

```python
fig = plt.figure(figsize= (15, 8))

ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ddqn_x = range(len(ddqn_reward_list[:5000]))
ax1.scatter(ddqn_x, ddqn_reward_list[:5000], s = 3, c = 'g')
ax1.set_title('DDQN')
ax1.set_xlabel('Episode')
ax1.set_ylabel('Reward')

ppo_x = range(len(ppo_reward_list[:5000]))
ax2.scatter(ppo_x, ppo_reward_list[:5000], s = 3, c = 'b')
ax2.set_title('PPO')
ax2.set_xlabel('Episode')
ax2.set_ylabel('Reward')

plt.tight_layout()
plt.show()
```

```python
ppo_3obs_log_txt = open('/workspace/V1_3obs/V1_3obs/output/ppo/ppo_log.txt', 'r')
ppo_4obs_log_txt = open('/workspace/V1_4obs/V1_4obs/output/ppo/ppo_log.txt', 'r')
ppo_5obs_log_txt = open('/workspace/V1_5obs/V1_5obs/output/ppo/ppo_log.txt', 'r')

ddqn_3obs_log_txt = open('/workspace/V1_3obs/V1_3obs/output/ddqn/ddqn_log.txt', 'r')
ddqn_4obs_log_txt = open('/workspace/V1_4obs/V1_4obs/output/ddqn/ddqn_log.txt', 'r')
ddqn_5obs_log_txt = open('/workspace/V1_5obs/V1_5obs/output/ddqn/ddqn_log.txt', 'r')

def make_reward_list(file_name):
    reward_list = []
    lines = file_name.readlines()
    for i in range(len(lines)):
        reward_list.append(float(lines[i].split(' ')[-1][:-2]))
    
    return reward_list

ppo_3obs_reward_list = make_reward_list(ppo_3obs_log_txt)
ppo_4obs_reward_list = make_reward_list(ppo_4obs_log_txt)
ppo_5obs_reward_list = make_reward_list(ppo_5obs_log_txt)

ddqn_3obs_reward_list = make_reward_list(ddqn_3obs_log_txt)
ddqn_4obs_reward_list = make_reward_list(ddqn_4obs_log_txt)
ddqn_5obs_reward_list = make_reward_list(ddqn_5obs_log_txt)

fig = plt.figure(figsize= (15, 12))

ax1 = fig.add_subplot(3, 2, 1)
ax2 = fig.add_subplot(3, 2, 2)
ax3 = fig.add_subplot(3, 2, 3)

ax4 = fig.add_subplot(3, 2, 4)
ax5 = fig.add_subplot(3, 2, 5)
ax6 = fig.add_subplot(3, 2, 6)

ddqn_x = range(len(ppo_3obs_reward_list[:5000]))
ax2.scatter(ddqn_x, ppo_3obs_reward_list[:5000], s = 3, c = 'g')
ax2.set_title('PPO - 3obs')
ax2.set_xlabel('Episode')
ax2.set_ylabel('Reward')

ppo_x = range(len(ppo_4obs_reward_list[:5000]))
ax4.scatter(ppo_x, ppo_4obs_reward_list[:5000], s = 3, c = 'b')
ax4.set_title('PPO - 4obs')
ax4.set_xlabel('Episode')
ax4.set_ylabel('Reward')

ppo_x = range(len(ppo_5obs_reward_list[:5000]))
ax6.scatter(ppo_x, ppo_5obs_reward_list[:5000], s = 3, c = 'r')
ax6.set_title('PPO - 5obs')
ax6.set_xlabel('Episode')
ax6.set_ylabel('Reward')



ddqn_x = range(len(ddqn_3obs_reward_list[:5000]))
ax1.scatter(ddqn_x, ddqn_3obs_reward_list[:5000], s = 3, c = 'g')
ax1.set_title('DDQN - 3obs')
ax1.set_xlabel('Episode')
ax1.set_ylabel('Reward')

ppo_x = range(len(ddqn_4obs_reward_list[:5000]))
ax3.scatter(ppo_x, ddqn_4obs_reward_list[:5000], s = 3, c = 'b')
ax3.set_title('DDQN - 4obs')
ax3.set_xlabel('Episode')
ax3.set_ylabel('Reward')

ppo_x = range(len(ddqn_5obs_reward_list[:5000]))
ax5.scatter(ppo_x, ddqn_5obs_reward_list[:5000], s = 3, c = 'r')
ax5.set_title('DDQN - 5obs')
ax5.set_xlabel('Episode')
ax5.set_ylabel('Reward')


print(len(([num for num in ppo_3obs_reward_list if num > 0])), np.mean([num for num in ppo_3obs_reward_list if num > 0]))
print(len(([num for num in ppo_4obs_reward_list if num > 0])), np.mean([num for num in ppo_4obs_reward_list if num > 0]))
print(len(([num for num in ppo_5obs_reward_list if num > 0])), np.mean([num for num in ppo_5obs_reward_list if num > 0]))
print(len(([num for num in ddqn_3obs_reward_list if num > 0])), np.mean([num for num in ddqn_3obs_reward_list if num > 0]))
print(len(([num for num in ddqn_4obs_reward_list if num > 0])), np.mean([num for num in ddqn_4obs_reward_list if num > 0]))
print(len(([num for num in ddqn_5obs_reward_list if num > 0])), np.mean([num for num in ddqn_5obs_reward_list if num > 0]))

plt.tight_layout()
plt.show()
```
