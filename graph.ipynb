```python
import matplotlib.pyplot as plt
```

```python
ppo_log_txt = open('/workspace/V0/output/ppo/ppo_log.txt', 'r')
q_log_txt = open('/workspace/V0/output/original/original_log.txt', 'r')
ddqn_log_txt = open('/workspace/V0/output/ddqn/ddqn_log.txt', 'r')
```

```python
def make_reward_list(file_name):
    reward_list = []
    lines = file_name.readlines()
    for i in range(len(lines)):
        reward_list.append(float(lines[i].split(' ')[-1][:-2]))
    
    return reward_list
```

```python
ppo_reward_list = make_reward_list(ppo_log_txt)
q_reward_list = make_reward_list(q_log_txt)
ddqn_reward_list = make_reward_list(ddqn_log_txt)
```

```python
fig = plt.figure(figsize= (15, 12))

ax1 = fig.add_subplot(3, 1, 1)
ax2 = fig.add_subplot(3, 1, 2)
ax3 = fig.add_subplot(3, 1, 3)


q_x = range(len(q_reward_list[:10000]))
ax1.scatter(q_x, q_reward_list[:10000], s = 3, c = 'r')
ax1.set_title('Q-Learning')
ax1.set_xlabel('Episode')
ax1.set_ylabel('Reward')

ddqn_x = range(len(ddqn_reward_list[:1500]))
ax2.scatter(ddqn_x, ddqn_reward_list[:1500], s = 3, c = 'g')
ax2.set_title('DDQN')
ax2.set_xlabel('Episode')
ax2.set_ylabel('Reward')

ppo_x = range(len(ppo_reward_list[:300]))
ax3.scatter(ppo_x, ppo_reward_list[:300], s = 3, c = 'b')
ax3.set_title('PPO')
ax3.set_xlabel('Episode')
ax3.set_ylabel('Reward')

plt.tight_layout()
plt.show()
```
